```python
from keras.models import load_model
import cv2
import numpy as np

model = load_model('trained_model.h5')

img = cv2.imread('Prediction/saf.jfif')

from tensorflow.keras.preprocessing import image
test_image = image.load_img('Prediction/saf.jfif',target_size=(64,64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image,axis=0)
result = model.predict(test_image)
print(result)
#training_set.class_indices

if result[0][0]==1:
    print('Saffron')
elif result[0][1]>=1:
    print('Daisy')
elif result[0][2]==1:
    print('Dandelion')
elif result[0][3]==1:
    print('Rose')
elif result[0][4]==1:
    print("SunFlower")
elif result[0][5]==1:
    print("Tulip")

import matplotlib.pyplot as plt
plt.imshow(img)
```
